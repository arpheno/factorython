```python
import pulp

def minimize_transportation_costs(distances, flows):
    n = len(distances)
    departments = range(n)

    # Create a linear programming problem
    model = pulp.LpProblem(name="Transportation_Cost_Minimization", sense=pulp.LpMinimize)

    # Decision variables: Position of each department on the line
    x = pulp.LpVariable.dicts("Department_Position", departments, lowBound=0, cat=pulp.LpContinous)

    # Binary variables to linearize the absolute value term
    y = pulp.LpVariable.dicts("Positive_Part", (departments, departments), cat=pulp.LpBinary)
    z = pulp.LpVariable.dicts("Negative_Part", (departments, departments), cat=pulp.LpBinary)

    # Objective function: Minimize total transportation costs
    model += pulp.lpSum(distances[i][j] * flows[i][j] * (y[i][j] + z[i][j]) for i in departments for j in departments)

    # Constraints: Each department must be assigned to a unique position on the line
    for i in departments:
        model += pulp.lpSum(x[i]) == 1

    # Constraints to enforce y[i][j] and z[i][j] as positive and negative parts of the absolute value
    for i in departments:
        for j in departments:
            model += x[i] - x[j] <= y[i][j] * n
            model += x[j] - x[i] <= z[i][j] * n

    # Solve the LP problem
    model.solve()

    if pulp.LpStatus[model.status] == "Optimal":
        # Return the optimal positions of departments and the total transportation cost
        optimal_positions = [x[i].varValue for i in departments]
        total_transportation_cost = pulp.value(model.objective)
        return optimal_positions, total_transportation_cost
    else:
        return None, None

# Example usage (replace example distance matrix and flow matrix with actual data)
if __name__ == "__main__":
    distances = [
        [0, 3, 6],
        [3, 0, 4],
        [6, 4, 0]
    ]
    flows = [
        [0, 50, 30],
        [50, 0, 20],
        [30, 20, 0]
    ]

    positions, total_cost = minimize_transportation_costs(distances, flows)
    
```
